# Plotting on Map using GeoPandas

The goal of GeoPandas is to make working with geospatial data in python easier. It extends the datatypes used by pandas to allow spatial operations on geometric types. Geometric operations are performed by shapely and geopandas uses descartes and matplotlib for plotting as well as fiona for file access. 
Source: [geopands.org](https://geopandas.org)

Conda package manager can be used to install geopandas and all its dependencies - <br>

GeoPandas implements two main data structures, <b>GeoSeries</b> and <b>GeoDataFrame</b> that are subclasses of pandas Series and DataFrame respectively. Each entry in the GeoSeries is a set of shapes corresponding to one observation. The shapes (that are actually shapely objects) can be points, lines or polygons. <br>
A GeoDataFrame is a data structure that contains a GeoSeries, ie., it will always contain one GeoSeries column that holds the GeoDF's "geometry". Any method(function) applied to the DF will be applied to the "geometry" column. <br>
A GeoDF may contain other columns with geometrical (shapely) objects, but only one column can be the active geometry at a time.

In this tutorial, I will use the USA shape file downloaded from US Census Bureau's Geographic Encoding and Referencing Database. [Click to download the link.](https://catalog.data.gov/dataset/tiger-line-shapefile-2017-nation-u-s-current-state-and-equivalent-national) <br>
First, lets import all necessary libraries

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
#read the shape file
usa_states = gpd.read_file('tl_2017_us_state/tl_2017_us_state.shp')
usa_states.head()
```

```python
usa_states.dtypes
```

```python
type(usa_states)
```

We can see that usa_states is GeoDataFrame and the geometry column in the DF is of type geometry where each datapoint is of type polygon/multi-polygon. Lets check to see if it is the active geometry column and then plot the map.

```python
usa_states.geometry.name
```

```python
#Basic plot - random color
usa_states.plot(figsize=(10,8))
```

The presence of Alaska, Hawaii, PR and various islands in the above map make it impossible to view it in higher resolution. One easy way to fix this is to remove rows with spatial data for those states.

```python
usa_states.sort_values(by='STATEFP')
usa_contiguous = usa_states[usa_states['STATEFP'].astype(int) < 60].copy()
idx = usa_contiguous[(usa_contiguous['NAME'] == 'Alaska') | (usa_contiguous['NAME'] == 'Hawaii')].index
usa_contiguous.drop(index = idx, inplace =True)
```

```python
usa_contiguous.plot(figsize = (10,8))
```

By performing simple pandas operation, we were able to remove all rows except the contiguous states. <br>
Any option one can pass to pyplot in matplotlib can be passed to plot() method

```python
#changing color to reds, linewidth to 0.5 and linecolor to black
usa_contiguous.plot(cmap = 'Reds', linewidth = 0.5, edgecolor = 'black', figsize=(12,8))
#deleting axis as they don't give us any information
plt.axis('off')
```

The colors above are randomly filled. We can use the 'column' parameter to fill the color by certain column. Below we are using the 'REGION' column from the GeoPandas DF.

```python
usa_contiguous.plot(column = 'REGION', cmap = 'Reds', linewidth = 0.5, edgecolor = 'black', figsize=(12,8))
plt.axis('off')
```

We can use the 'column' feature to easily create Choropleth maps (maps where the color of each shape visualizes information across a national or global focus). In this case, I will use the data from my 'OUTBREAK-OF-COVID-19' folder to plot the number of cases across the states. The file is stored locally as 'COVID_USA_STATES_YYYY-MM-DD.csv'. The date indicates when the file was current.

```python
covid_usa = pd.read_csv('COVID_USA_States_2019-04-19.csv')
covid_usa.head()
```

We will merge the two dataframes on State Names to get one GeoPandas DF. Keep in mind that we should merge the <code>covid_usa</code> DF to <code>usa_contiguous</code> DF so that the resulting DF is a GeoPandas DF.

```python
merged_df = usa_contiguous.merge(covid_usa, how ='left', left_on='NAME', right_on='Province_State')
print(type(merged_df)) #confirming type
print(merged_df.geometry.name) #confirming geometry column
merged_df.head()
```

Lets clean up the merged DF by removing some duplicate columns as well as columns that are not required for this tutorial.

```python
cols = ['REGION', 'DIVISION', 'STATENS', 'GEOID', 'STUSPS', 'LSAD', 'MTFCC', 'FUNCSTAT', 'Province_State', 'Country']
merged_df.drop(columns=cols, inplace = True)
```

```python
merged_df.head()
```

Lets plot a choropleth map using <code>Confirmed Count</code> as the column

```python
fig, ax = plt.subplots(1,1,figsize=(15,8))
merged_df.plot(column = 'Confirmed Count', cmap = 'Blues', 
               linewidth=0.5, edgecolor='0.7', 
               legend = True, legend_kwds = {'label': 'Confirmed Cases',
                                            'orientation': 'horizontal'},
               ax=ax)
```

The above plot does not give us a lot of information about the cases in each state. This is because the column's range is large and the high counts are localized to very small number of states. The default scaling can be changed using the scheme option of the plot. The 'scheme' option requires you to have 'mapclassify' installed using either of two commands - <br>
<code> pip install mapclassify</code> <br> <code> conda install -c conda -forge mapclassify</code> <br>

```python
fig, ax = plt.subplots(1,1,figsize=(15,8))
merged_df.plot(column = 'Confirmed Count', cmap = 'Blues', 
               linewidth=0.5, edgecolor='0.7', 
               scheme = 'Percentiles', 
               legend = True, legend_kwds = {'loc': 'lower right'},
               ax=ax)
ax.axis('off')
fig.suptitle('Confirmed Cases')
```

We can also calculate the centroids(center) for each polygon and set that to geometry. Lets see how the plot changes when we change the geometry column.

```python
merged_df['centroid'] = merged_df.centroid
merged_df = merged_df.set_geometry('centroid')
merged_df.geometry.name
```

```python
merged_df.plot(column = 'Confirmed Count', scheme = 'Percentiles', legend = True, figsize=(15,12))
plt.axis('off')
```

The plot above is similar to a scatter plot on the US map. The location is the based on centroid values and the color of the points are based on the 'confirmed count' values. However, we loose the area and boundry locations in this case.

```python
#change back the geometry to geomtery
merged_df = merged_df.set_geometry('geometry')
merged_df.geometry.name
```

I will try to create a 'bubble/scatter plot' (similar to plotly's bubble plot) where the size of the marker represents the number of confirmed counts in each state. I will position the marker at the center(centroid) of each polygon. 

```python
fig, ax = plt.subplots(1,1, figsize=(20,15))
merged_df = merged_df.set_geometry('geometry')
merged_df.plot(ax=ax, color = 'gray')

#change geometry to centroid for scatter plot
merged_df = merged_df.set_geometry('centroid')
markersize = ((merged_df['Confirmed Count'] - merged_df['Confirmed Count'].min()) / (merged_df['Confirmed Count'].max() - merged_df['Confirmed Count'].min()))*100
merged_df.plot(ax=ax, column = 'Confirmed Count', 
               markersize = markersize*50, scheme = 'Percentiles', 
               color = 'Red', alpha = 0.5, 
               legend=True)
```
